```python
import seaborn as sn
corr = df.corr()

sn.heatmap(corr, annot=True, cmap='Blues')
```

```python
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df.Outcome
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8)
```

```python
print('The shape of X_train:', X_train.shape)
print('The shape of X_test:', X_test.shape)
print('The shape of y_train:', y_train.shape)
print('The shape of y_test:', y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)
y_hat
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Lets define the sigmoid function now and do maths with hand
import math
def sigmoid(x):
    return 1/(1 + math.exp(-x))

def prediction_function(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age):
    z = 0.12*Pregnancies + 0.04*Glucose -0.02*BloodPressure + 0.003*SkinThickness - 0.002*Insulin + 0.079*BMI + 0.663*DiabetesPedigreeFunction + 0.006*Age -7.475
    y = sigmoid(z)
    return y
    
```
